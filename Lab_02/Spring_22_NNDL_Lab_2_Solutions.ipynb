## Theory
Perceptron Learning Law:
 

*   The perceptron model have the ability to learn the weights and update it iteratively.
*   Activation value is $w^{T}x$ and activation function is hard limiting function.
*   Perceptron model learns the weights by minimizing error as close to 0.




*   $δ = b − s$, where b is target value, s in predicted value.
*   Weight update equation is $w^{'}_{i} = w_{i} + Δw_{i}$.
*   $Δw_{i} = ηδa_{i}$

```python
import numpy as np
import matplotlib.pyplot as plt

def GenerateData(No_training,mean,variance):

  data = np.random.multivariate_normal(mean,variance,No_training)
  ## Write your code here

  return data
```

```python
## Define Mean and Variance for all the data points
N = 100 

mean1 = np.array([0,0])
mean2 = np.array([0,10])
mean3 = np.array([10,0])
mean4 = np.array([10,10]) 

var = np.array([[1,0],[0,1]])
```

## Algorithm


*   We are passing all the inputs to two perceptron with Hard limiting activation function.
*   The perceptron will generate values for LSB and MSB.
*   From LSB and MSB the Labels can be calculated as follows $labels = 2*MSB + LSB$
  \begin{array}{|c|c|c|}\hline\
  \textbf{b2} & \textbf{b1} & \textbf{Label = 2*MSB+LSB} & \textbf{Label} \\ \hline\
  0 & 0 & 2*0+0 & 0   \\ \hline\
  0 & 1 & 2*0+1 & 1   \\ \hline\
  1 & 0 & 2*1+0 & 2   \\ \hline\
  1 & 1 & 2*1+1 & 3   \\ \hline
  \end{array}

```python
## Define Target Labels
trn_lbl=label
b1 = np.zeros((trn_lbl.shape[0],1))
b2 = np.zeros((trn_lbl.shape[0],1))
for i in range(trn_lbl.shape[0]):
  ## LSB
  if (np.mod(trn_lbl[i],2)==0):
    b1[i]=0
  else:
    b1[i]=1
  ## MSB
  if trn_lbl[i]<2:
    b2[i]=0
  else:
    b2[i]=1
# print(b1[i],b2[i])
# print(b1)
# print(b2)
# Data Augmentation
bias = np.ones(trn_lbl.shape[0])
data_aug = np.column_stack((bias,data))
```

## Perceptron Learning Function

```python
def perceptron_learning_loop(data_aug, b1, b2):

  ## Write your code here

  lr=0.5
  w1=np.random.uniform(-1,1,(data_aug.shape[1],1))
  w2=np.random.uniform(-1,1,(data_aug.shape[1],1))
  # print(w.shape)
  error1=[]
  error2=[]
  for itr in range(250):
    ## Activation value
    ## For LSB
    y_est1=data_aug @ w1
    # sign function
    y_pred1=np.ones(y_est1.shape)
    y_pred1[np.where(y_est1<=0)]=0
    # gradient based perceptron learning law
    err1 = (b1 - y_pred1)
    error1.append(np.linalg.norm(err1))
    # print(np.shape(err))
    del_w1 = lr*np.dot(data_aug.T,err1)
    w1 = w1+del_w1
    #print(w)

    ## For MSB
    y_est2=data_aug @ w2
    # sign function
    y_pred2=np.ones(y_est2.shape)
    y_pred2[np.where(y_est2<=0)]=0
    # gradient based perceptron learning law
    err2 = (b2 - y_pred2)
    error2.append(np.linalg.norm(err2))
    # print(np.shape(err))
    del_w2 = lr*np.dot(data_aug.T,err2)
    w2 = w2+del_w2
    ###########################
    miss1=(np.sum(np.abs(b1-y_pred1))) # no of miss classification
    miss2=(np.sum(np.abs(b2-y_pred2)))  

    # for ploting
    if (np.remainder(itr,50)==0):
      print('After iteration=',itr)
      print('miss classified=',miss1)
      print('miss classified=',miss2)
      x_dc=np.linspace(np.min(data_aug[:,1]),np.max(data_aug[:,1]),1000)
      leq1=((-w1[0]-(w1[1]*x_dc))/(w1[2]))
      leq2=((-w2[0]-(w2[1]*x_dc))/(w2[2]))
      plt.figure()
      plt.plot(x1[:,0],x1[:,1],'.',color='r')
      plt.plot(x2[:,0],x2[:,1],'.',color='y')
      plt.plot(x3[:,0],x3[:,1],'.',color='g')
      plt.plot(x4[:,0],x4[:,1],'.',color='b')
      plt.plot(x_dc,leq1)
      plt.plot(x_dc,leq2)
      plt.title('Iteration= %d' % (itr))
  plt.figure()
  plt.plot(error1)
  plt.title('error vs Iteration')

  plt.figure()
  plt.plot(error2)
  plt.title('error vs Iteration')
  return w1,w2,y_pred1,y_pred2

w1,w2,s1,s2 = perceptron_learning_loop(data_aug,b1,b2) ## Learn LSB weights
# w2,s2 = perceptron_learning_loop(data_aug,b2) ## Learn MSB weights
```

## Testing the model 

```python
## Testing Data generation
N = 100 

mean1 = np.array([0,0])
mean2 = np.array([0,10])
mean3 = np.array([10,0])
mean4 = np.array([10,10]) 

var = np.array([[1,0],[0,1]])

x1 = GenerateData(N,mean1,var)
y1 = np.zeros(N)
x2 = GenerateData(N,mean2,var)
y2 = np.ones(N)
x3 = GenerateData(N,mean3,var)
y3 = 2*np.ones(N)
x4 = GenerateData(N,mean4,var)
y4 = 3*np.ones(N)

data = np.concatenate((x1,x2,x3,x4))
# print(data)
label = np.concatenate((y1,y2,y3,y4))
# print(label)
##########################################################
## Converting labels to binary equivalent
trn_lbl=label
b1 = np.zeros((trn_lbl.shape[0],1))
b2 = np.zeros((trn_lbl.shape[0],1))
for i in range(trn_lbl.shape[0]):
  if (np.mod(trn_lbl[i],2)==0):
    b1[i]=0
  else:
    b1[i]=1
  if trn_lbl[i]<2:
    b2[i]=0
  else:
    b2[i]=1
# print(b1[i],b2[i])
# print(b1)
# print(b2)


### Data Augmentation
bias = np.ones(trn_lbl.shape[0])
data_aug = np.column_stack((bias,data))
```

##Predicting labels over test samples and calculating accuracy

```python
def predValue(data_aug,w):
  y_est=data_aug @ w
  # sign function
  y_pred=np.ones(y_est.shape)
  y_pred[np.where(y_est<=0)]=0
  return y_pred

s1 = predValue(data_aug, w1) ## Predicted LSB
s2 = predValue(data_aug, w2) ## Predicted MSB


predicted = np.zeros((trn_lbl.shape[0],1))
correct=0
for i in range(trn_lbl.shape[0]):
  predicted[i] = 2*s2[i]+s1[i]
  if (s1[i]==b1[i] and s2[i]==b2[i]): ## The labels are matched with predicted label to calculating accuracy.
    correct = correct+1
Accuracy = (correct/(trn_lbl.shape[0]))*100 ## Percentage Accuracy
print("Accuracy", Accuracy)
```
